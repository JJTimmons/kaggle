```python
import pandas as pd
import seaborn as sns

data = pd.read_csv("Admission_Predict_Ver1.1.csv")
high = data.loc[data["University Rating"] == 5]

high = high.drop(labels=["Serial No.", "TOEFL Score", "University Rating"], axis=1)
high.head()
```

```python

high_corr = high.corr()

cmap = sns.diverging_palette(220, 10, as_cmap=True)

sns.heatmap(high_corr,cmap=cmap, center=0,
            square=True, linewidths=.5, annot=True)
```
