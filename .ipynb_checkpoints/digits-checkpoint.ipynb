```python
import pandas as pd

data = pd.read_csv("train.csv")
data.head()
```

```python
import numpy as np
import math
from keras.utils import to_categorical

labels = []
images = []
for i, row in data.iterrows():
    labels.append(row[0])
    images.append(row[1:])

y = np.array(labels)
x = np.array(images)

scale = np.max(x)
x = np.divide(x, scale)

y = to_categorical(y, num_classes=10)

print(y.shape)
```

```python
import keras

from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout, Masking, Embedding, Flatten, Conv1D
from sklearn import pipeline


print(x.shape)

model = Sequential()
model.add(Dense(256, input_shape=(x.shape[1],), activation="relu"))
model.add(Dropout(0.15))
model.add(Dense(256, activation="relu"))
model.add(Dropout(0.15))
model.add(Dense(10, activation="softmax"))
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

print(model.summary())

model.fit(x=x, y=y, validation_split=0.1, verbose=1, epochs=15)
```

```python
test = pd.read_csv("test.csv")
test.head()

images = []
for i, row in test.iterrows():
    images.append(row[:])

x = np.array(images)

scale = np.max(x)
x = np.divide(x, scale)
```

```python
predictions = model.predict_classes(x)

with open("guesses.csv", "w") as out:
    out.write("ImageId,Label\n")
    for i, pred in enumerate(predictions):
        out.write(f"{i+1},{pred}\n")
```
